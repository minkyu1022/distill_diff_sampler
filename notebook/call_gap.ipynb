```python
import matplotlib.pyplot as plt

# === Plot styling ===
plt.rcParams.update({
    'font.family': 'Serif',
    'font.size': 12,
    'axes.labelsize': 10,
    'axes.titlesize': 10,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    'figure.dpi': 300,
    'lines.linewidth': 1.5,
})

# === Data points 정의 ===
data = {
    'Ours': [
        (10000, 46.089),
        (15000, 27.982),
        (19900, 22.026),
        (40000, 20.071),
        (45000, 19.652),
    ],
    'PIS': [
        (10000, 107.214),
        (15000, 76.355),
        (20000, 68.825),
        (25000, 60.507),
        (30000, 57.796),
        (35000, 52.167),
        (40000, 56.481),
        (45000, 48.247),
        (50000, 51.908),
        (55000, 42.237),
        (60000, 40.012),
        (65000, 32.028),
        (70000, 29.10),
        (75000, 33.838),
        (80000, 32.263),
    ],
    # 필요하면 여기에 다른 메소드 추가
}

# === Style map ===
style_map = {
    'PIS':  {'marker': 'X', 'color': 'tab:orange'},
    'Ours': {'marker': 'o', 'color': 'blue'},
    # 'SubTB+LP':      {'marker': '^', 'color': 'tab:cyan'},
    # 'TB+LP':         {'marker': 's', 'color': 'red'},
    # 'TB+Expl+LS+LP': {'marker': 'D', 'color': 'magenta'},
}

# === Plotting ===
fig, ax = plt.subplots(figsize=(4, 2.4))

for method, pts in data.items():
    # x, y 분리
    xs, ys = zip(*pts)
    ax.scatter(
        xs, ys,
        marker=style_map[method]['marker'],
        color=style_map[method]['color'],
        s=150,
        alpha=0.8,
        label=method
    )

# Labels and styling
ax.set_xlabel('Energy Calls')
ax.set_ylabel('EUBO - ELBO')
ax.set_xlim(right=9e4)
ax.set_ylim(bottom=12, top=120)
ax.axhline(200, color='gray', linestyle='--', alpha=0.7)
ax.legend()
ax.grid(True, linestyle=':', linewidth=0.8)

plt.tight_layout()
plt.show()

# 원하는 경로에 저장
fig.savefig('../figure/lj55_trade_off.png', format='png', bbox_inches='tight')
fig.savefig('../figure/lj55_trade_off.pdf', format='pdf', bbox_inches='tight')
```
