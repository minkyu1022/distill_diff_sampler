```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FuncFormatter  # 추가

# === Plot styling ===
plt.rcParams.update({
    'font.family': 'Serif',
    'font.size': 12,
    'axes.labelsize': 10,
    'axes.titlesize': 10,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    'figure.dpi': 300,
    'lines.linewidth': 1.5,
})

# === Data points 정의 ===
data = {
    'SGDS': [
        (10000, 51.475),
        (15000, 38.201),
        (19900, 35.506),
        # (40000, 20.071),
        (45000, 33.014),
    ],
    'PIS': [
        (10000, 92.792),
        (15000, 75.138),
        (20000, 60.890),
        # (25000, 60.507),
        (30000, 52.796),
        # (35000, 52.167),
        (40000, 53.275),
        # (45000, 48.247),
        (50000, 51.414),
        # (55000, 42.237),
        (60000, 43.526),
        # (65000, 32.028),
        (70000, 45.201),
        # (75000, 33.838),
        # (80000, 32.263),
    ],
    # 필요하면 여기에 다른 메소드 추가
}

# === Style map ===
style_map = {
    'PIS':  {'marker': 'X', 'color': 'tab:orange'},
    'SGDS': {'marker': 'o', 'color': 'blue'},
}

# === Plotting ===
fig, ax = plt.subplots(figsize=(4, 3))

for method, pts in data.items():
    # x, y 분리
    xs, ys = zip(*pts)
    ax.scatter(
        xs, ys,
        marker=style_map[method]['marker'],
        color=style_map[method]['color'],
        s=50,
        alpha=0.8,
        label=method
    )

# Labels and styling
ax.set_xlabel(r'Energy Calls [$\times 10^4$]')
ax.set_ylabel('EUBO - ELBO')
ax.set_xlim(right=7.5e4)
ax.set_ylim(bottom=12, top=120)
ax.axhline(200, color='gray', linestyle='--', alpha=0.7)
ax.set_xticks([20000, 40000, 60000])
ax.set_xticklabels(['2', '4', '6'])

ax.legend()
leg = ax.get_legend()
for text in leg.get_texts():
    if text.get_text() == 'SGDS':  # "우리꺼"로 표시하고 싶다면
        text.set_text('SGDS')
        text.set_fontstyle('italic')
ax.grid(True, linestyle=':', linewidth=0.8)

plt.tight_layout()
plt.show()

# 원하는 경로에 저장
fig.savefig('../figure/lj55_trade_off.png', format='png', bbox_inches='tight')
fig.savefig('../figure/lj55_trade_off.pdf', format='pdf', bbox_inches='tight')
```
