```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    'font.family': 'serif', 
    'font.size': 14,
    'axes.labelsize': 10,
    'axes.titlesize': 10,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 8,
    'figure.dpi': 300,
    'lines.linewidth': 1.5,
    'axes.grid': True,
})

def draw_interatomic_dist_histogram(ax, interatomic_dist, name=None, bins=100):
    ax.set_xlabel("Interatomic distances")
    ax.set_ylabel("Density")
    ax.grid(True, linestyle="--", alpha=0.7)

    return ax.hist(
        interatomic_dist,
        bins=bins,
        range=(0, 6),
        alpha=0.5,
        label=name,
        density=True,
        histtype="step",
        linewidth=4,
    )
    
def plot_dist(name, dist_dict):
    fig, ax = plt.subplots(1, 1, figsize=(6, 6))

    for method, interatomic_dist in dist_dict.items():
        draw_interatomic_dist_histogram(ax, interatomic_dist, method)

    ax.legend()
    plt.tight_layout()
    plt.savefig(f'../figure/{name}_interatomic_distances.png', dpi=300)

name = 'lj13'
dist_dict = {
    'GT': np.load(f'../result/{name}/gt_distances.npy'),
    'Teacher': np.load(f'../result/{name}/teacher_distances.npy'),
    'Ours': np.load('../result/2025-04-29_16:24:10/distances.npy'),
    'PIS+LP': np.load('../result/2025-04-29_16:24:11/distances.npy'),
    # 'SubTB+LP': np.load('../result/2025-04-29_16:24:12/distances.npy'),
    'TB+Expl+LS+LP': np.load('../result/2025-04-29_16:24:13/distances.npy'),
    'TB+Expl+LS': np.load('../result/2025-04-29_16:24:14/distances.npy'),
    'TB+LP': np.load('../result/2025-04-29_16:24:15/distances.npy'),
    'TB+LS': np.load('../result/2025-04-29_16:24:16/distances.npy'),
    'TB': np.load('../result/2025-04-29_16:24:17/distances.npy'),
}

plot_dist(name, dist_dict)
```
