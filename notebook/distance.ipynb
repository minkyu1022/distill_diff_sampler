```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    'font.family': 'serif', 
    'font.size': 14,
    'axes.labelsize': 10,
    'axes.titlesize': 10,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 8,
    'figure.dpi': 300,
    'lines.linewidth': 1.5,
})

def draw_interatomic_dist_histogram(ax, interatomic_dist, lower, upper, name=None, bins=100):
    ax.set_xlabel("Interatomic distances")
    ax.set_ylabel("Density")
    ax.grid(True, linestyle="--", alpha=0.7)

    return ax.hist(
        interatomic_dist,
        bins=bins,
        range=(lower, upper),
        alpha=0.5,
        label=name,
        density=True,
        histtype="step",
        linewidth=4,
    )
    
def plot_dist(name, dist_dict, lower, upper):
    plt.clf()
    plt.close()
    fig, ax = plt.subplots(1, 1, figsize=(6, 3))

    for method, interatomic_dist in dist_dict.items():
        draw_interatomic_dist_histogram(ax, interatomic_dist, lower, upper, method)

    ax.legend()
    plt.tight_layout()
    plt.savefig(f'../figure/{name}_interatomic_distances.png')

name = 'lj13'
dist_dict = {
    'GT': np.load(f'../result/2025-05-06_12:23:44/dist/GT_4000.npy'),
    # 'Teacher': np.load(f'../result/2025-05-06_12:23:44/dist/Teacher_4000.npy'),
    # 'Ours': np.load('../result/2025-05-06_12:23:44/dist/Student_4000.npy'),
    # 'TB': np.load('../result/2025-05-06_12:23:46/dist/Student_4000.npy'),
    # 'PIS': np.load('../result/2025-05-06_12:23:48/dist/Student_4000.npy'),
    # 'TB+LS': np.load('../result/2025-05-06_12:23:50/dist/Student_4000.npy'),
    # 'TB+Expl+LS': np.load('../result/2025-05-06_12:23:52/dist/Student_4000.npy'),
    'iDEM': np.load(f'../result/iDEM_{name}/dist/Student.npy').reshape(-1),
    # 'TB+Expl+LP': np.load('../result/2025-05-05_21:50:39/dist/Student_4000.npy'),
    # 'TB+Expl+LS+LP': np.load('../result/2025-05-05_21:50:41/dist/Student_4000.npy'),
    # 'TB+LP': np.load('../result/2025-05-05_21:50:45/dist/Student_4000.npy'),
    # 'TB+LS+LP': np.load('../result/2025-05-05_21:50:47/dist/Student_4000.npy'),
}

lower = 0
upper = 6

# name = 'lj55'
# dist_dict = {
#     'GT': np.load(f'../result/2025-05-06_12:23:44/dist/GT_4000.npy'),
#     # 'Teacher': np.load(f'../result/2025-05-06_12:23:44/dist/Teacher_4000.npy'),
#     # 'Ours': np.load('../result/2025-05-06_12:23:44/dist/Student_4000.npy'),
#     # 'TB': np.load('../result/2025-05-06_12:23:46/dist/Student_4000.npy'),
#     # 'PIS': np.load('../result/2025-05-06_12:23:48/dist/Student_4000.npy'),
#     # 'TB+LS': np.load('../result/2025-05-06_12:23:50/dist/Student_4000.npy'),
#     # 'TB+Expl+LS': np.load('../result/2025-05-06_12:23:52/dist/Student_4000.npy'),
#     'iDEM': np.load(f'../result/iDEM_{name}/dist/Student.npy').reshape(-1),
#     # 'TB+Expl+LP': np.load('../result/2025-05-05_21:50:39/dist/Student_4000.npy'),
#     # 'TB+Expl+LS+LP': np.load('../result/2025-05-05_21:50:41/dist/Student_4000.npy'),
#     # 'TB+LP': np.load('../result/2025-05-05_21:50:45/dist/Student_4000.npy'),
#     # 'TB+LS+LP': np.load('../result/2025-05-05_21:50:47/dist/Student_4000.npy'),
# }

# lower = 0
# upper = 6

# name = 'aldp'
# dist_dict = {
#     'GT': np.load(f'../result/2025-05-05_21:50:37/dist/GT_0.npy'),
#     'Teacher': np.load(f'../result/2025-05-05_21:50:37/dist/Teacher_0.npy'),
#     'Ours': np.load('../result/2025-05-05_21:50:37/dist/Student_0.npy'),
#     'MLE': np.load('../result/2025-05-05_21:50:37/dist/Student_0.npy'),
#     'FAB': np.load(f'../result/FAB_{name}/dist/Student.npy'),
# }

# lower = 0
# upper = 2
plot_dist(name, dist_dict, lower, upper)
```
