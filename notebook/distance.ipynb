```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    'font.family': 'serif',
    'font.size': 14,
    'axes.labelsize': 10,
    'axes.titlesize': 10,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 8,
    'figure.dpi': 300,
    'lines.linewidth': 1.5,
})

palette_dict = {
    "Reference": "#808080",     # Gray
    "Ours": "#1F77B4",     # Blue
    "PIS": "#FF7F0E",     # Orange
    "iDEM": "#2CA02C",     # Green
    "TB+Expl+LS": "#9467BD",     # Purple
    "MLE": "#8C564B",     # Brown
    "FAB": "#E377C2",     # Pink
}

def draw_interatomic_dist_histogram(ax, interatomic_dist, lower, upper, name=None, bins=100):
    ax.set_xlabel("Interatomic distances")
    # ax.set_ylabel("Density")
    ax.grid(True, linestyle="--", alpha=0.7)
    return ax.hist(
        interatomic_dist,
        bins=bins,
        range=(lower, upper),
        alpha=0.5,
        label=name,
        density=True,
        histtype="step",
        linewidth=4,
        color=palette_dict[name],
    )
    
def plot_dist(name, dist_dict, lower, upper):
    plt.clf()
    plt.close()
    fig, ax = plt.subplots(1, 1, figsize=(4, 2))

    for method, interatomic_dist in dist_dict.items():
        draw_interatomic_dist_histogram(ax, interatomic_dist, lower, upper, method)

    ax.legend()
    plt.tight_layout()
    plt.savefig(f'../figure/{name}_interatomic_distances.png', format='png', bbox_inches='tight')
    plt.savefig(f'../figure/{name}_interatomic_distances.pdf', format='pdf', bbox_inches='tight')

# name = 'lj13'
# dist_dict = {
#     'Reference': np.load(f'../assets/lj13/ref/distances.npy'),
#     'Ours': np.load('../assets/lj13/ours/distances.npy'),
#     'PIS': np.load('../assets/lj13/pis/distances.npy'),
#     'iDEM': np.load(f'../assets/lj13/idem/distances.npy'),
# }

# lower = 0
# upper = 6

# name = 'lj55'
# dist_dict = {
#     'Reference': np.load(f'../assets/lj55/ref/distances.npy'),
#     'Ours': np.load('../assets/lj55/ours/distances.npy'),
#     'PIS': np.load('../assets/lj55/pis/distances.npy'),
#     'TB+Expl+LS': np.load(f'../assets/lj55/tb_expl_ls/distances.npy'),
#     # 'iDEM': np.load(f'../assets/lj55/idem/distances.npy'),
# }

# lower = 0
# upper = 6

name = 'aldp'
dist_dict = {
    'Reference': np.load('../assets/aldp/ref/distances.npy'),
    # 'PIS': np.load(f'../assets/aldp/pis/distances.npy'),
    # 'TB+Expl+LS': np.load(f'../assets/aldp/tb_expl_ls/distances.npy'),
    'Ours': np.load('../assets/aldp/ours/distances.npy'),
    'MLE': np.load('../assets/aldp/mle/distances.npy'),
    # 'FAB': np.load(f'../assets/aldp/fab/distances.npy'),
}

lower = 0
upper = 1
plot_dist(name, dist_dict, lower, upper)
```
